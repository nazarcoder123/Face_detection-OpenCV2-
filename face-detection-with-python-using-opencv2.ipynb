## Face Detection with Python using OpenCV

## 1.Import OpenCV library

```python
import cv2
```

## 2.Using the pre-trained classifier from OpenCV

```python
face_cascade = cv2.CascadeClassifier("/kaggle/working/haarcascade_frontalface_default.xml") #Load the CasCade
```

## 3.Upload the imput image

```python
img = cv2.imread("/kaggle/input/magnus-img/images.jpeg") #Read the input image
```

## 4.Perform face detection

```python
faces = face_cascade.detectMultiScale(img, 1.1, 4) #Detect faces
```

## 5.Drawing rectangle around the faces

```python
# Draw rectangle around the faces

for(x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255, 0, 0), 2)
```

## 6.Exporting the image file

```python
#Export the result
cv2.imwrite("face_detected_magnus.png", img)
print("photo successfully exported!")
```
